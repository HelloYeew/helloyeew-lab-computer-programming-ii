# Titanic (5 points)

Import titanic passenger data

Note: please upload `titanic2.csv` file using File Explorer tool on your left

**This is different from `titanic.csv` from the previous lab.**

image.png

```python
titanic = pd.read_csv('titanic2.csv')
titanic
```

Set `PassengerId` as the index

Hint: use `set_index()` method

```python
titanic.set_index('PassengerId')
```

Visualize a pie chart representing the gender proportion

example: image.png

```python
male = titanic[titanic["Sex"] == "male"].count()["Sex"]
female = titanic[titanic["Sex"] == "female"].count()["Sex"]
sex_df = pd.DataFrame({'Sex': [male, female]}, index=['Male', 'Female'])
plot = sex_df.plot.pie(y='Sex', autopct="%1.1f%%")
# sex_df
```

Visualize a scatter plot with the Fare and the Age, each dot colored by gender

example: image.png

```python
ax = titanic[titanic['Sex'] == 'male'].plot.scatter(x='Age', y='Fare', color="Purple", label='male')
ax = titanic[titanic['Sex'] == 'female'].plot.scatter(x='Age', y='Fare', color="Green", label='female',ax=ax)
```

Visualize a histogram with the Fare

example: image.png

```python
fare_his = titanic['Fare'].hist()
fare_his
```

Visualize a bar plot representing survival rates in percent grouped by the passenger class

```python
all_people = titanic['Pclass'].value_counts()
survived_people = titanic.loc[titanic["Survived"] == 1, 'Pclass'].value_counts()
percentage = survived_people / all_people * 100
percentage.plot.bar(xlabel = "Class", ylabel="Rate of survival")
```

# Apple Stock (10 points)

Import dataset from `apple_stock_1980_2014.csv`

```python
apple_stock = pd.read_csv('apple_stock_1980_2014.csv')
apple_stock
```

Print the type of each data column

```python
apple_stock.dtypes
```

Convert the Date column to a datetime type

```python
apple_stock['Date'] = pd.to_datetime(apple_stock["Date"])
apple_stock
```

Set the date as the index

```python
apple_stock.set_index('Date')
```

Check if there is any duplicate dates

```python
apple_stock["Date"].is_unique
```

Currently, the index is from the most recent date. Sort the entries so that the oldest date comes first.

```python
apple_stock.sort_values(by='Date')
```

Find the last business day of each month using `resample('BM')`

```python
apple_stock.set_index('Date', inplace=True)
apple_stock.resample('BM').max()
```

Compute the difference in days between the first and the last day

```python
(apple_stock.index[-1] - apple_stock.index[0]).days
```

Compute the number of months in the data we have

```python
apple_stock.index.month.value_counts().sort_index()
```

Plot the 'Adj Close' value.

```python
apple_stock['Adj Close'].plot()
```
