# Give an example code to sum a small array that runs faster than `np.sum`?

Use `%timeit` to prove that your **sum** is faster than `np.sum`

```python
a = np.arange(10)
%timeit np.sum(a)
```

```python
def sum_check():
  q = 0
  for i in a:
    q += i
  return q
%timeit sum_check()
```

# Write a code to plot both sine and cosine curves in the same plot with proper legend.

Use the code below as your starting point

```python
import matplotlib.pyplot as plt
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
y_cos = np.cos(x)
# Plot the points using matplotlib
plt.plot(x, y)
plt.plot(x, y_cos)

plt.xlim(0, 3 * np.pi)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sinusoidal curve(s)')
plt.legend(['Sine', 'Cosine'])
```

# Subplots

You can display different plots in the same figure using the subplots function. Here is an example:

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
fig, axes = plt.subplots(nrows=2)
axes[0].plot(x, y_sin)
axes[0].set_xlim(0, 3 * np.pi)
axes[0].set_title('Sine')
axes[1].plot(x, y_cos)
axes[1].set_xlim(0, 3 * np.pi)
axes[1].set_title('Cosine')
fig.show()
```

# Plotting

Produce a plot of the functions $f(x)=e^{-x/10}\sin(\pi x)$ and $g(x)=xe^{−x/3}$ over the interval $[0, 10]$. Include labels for the x- and y-axes, and a legend explaining which line is which plot.

Hint:
*   `np.pi`
*   `np.sin()`
*   `np.exp()`



```python
x_plot = np.arange(0,10,0.1)
y_f = np.exp(-x_plot/10)*np.sin(np.pi*x_plot)
y_g = x_plot*(np.exp(-x_plot/3))
plt.plot(x_plot, y_f)
plt.plot(x_plot, y_g)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)', 'g(x)'])
```
